```julia
using Arrhenius
using ForwardDiff
using LinearAlgebra
using DiffEqSensitivity
using DifferentialEquations
using Sundials
using StatsPlots
using Distributions

using Random
using ProgressBars
using DelimitedFiles
using Plots, Printf, Profile
using Dierckx # for interpolation
```

```julia
gas = CreateSolution("../../mechanism/gri30_217.yaml")
ns = gas.n_species
nr = gas.n_reactions

P = one_atm;
T0 = 1200.0;
phi = 1.0;

datasize = 400;
tsteps = range(0, 2e-1, length=datasize);
tspan = [0.0, tsteps[end]*1.001];

abstol = 1e-9;
reltol = 1e-12;
```

```julia
uncertainty_data = readdlm("../../mechanism/gri30_217.txt");
UF = uncertainty_data[:,2];
```

```julia
@inbounds function dudt!(du, u, p, t)
    T = u[end]
    Y = @view(u[1:ns])
    
    mean_MW = 1. / dot(Y, 1 ./ gas.MW)
    ρ_mass = P / R / T * mean_MW
    X = Y2X(gas, Y, mean_MW)
    C = Y2C(gas, Y, ρ_mass)
    cp_mole, cp_mass = get_cp(gas, T, X, mean_MW)
    h_mole = get_H(gas, T, Y, X)
    S0 = get_S(gas, T, P, X)
    kp = exp.(p)
    qdot = wdot_func(gas.reaction, T, C, S0, h_mole; get_qdot=true) .* kp
    wdot = gas.reaction.vk * qdot
    Ydot = wdot / ρ_mass .* gas.MW
    Tdot = -dot(h_mole, wdot) / ρ_mass / cp_mass
    du .= vcat(Ydot, Tdot)
end
X0 = zeros(ns);
X0[species_index(gas, "CH4")] = phi
X0[species_index(gas, "O2")] = 2.0
X0[species_index(gas, "N2")] = 2.0 * 3.76
X0 = X0 ./ sum(X0);
Y0 = X2Y(gas, X0, dot(X0, gas.MW));
u0 = vcat(Y0, T0);
```

```julia
function get_Tcurve(prob, p; tfinal=1.0, saveat=[])
    sol = solve(prob, p=p, CVODE_BDF(), saveat=saveat, reltol=reltol, abstol=abstol);
    ts = sol.t
    pred = clamp.(Array(sol), 0, Inf);
    return ts, pred
end
function diffMax(x, y)
    pos = argmax(diff(y)./diff(x))
    return pos, x[pos]
end
function curvMax(x, y; n=5, N=100)
    l = length(x) - 1;
    pos, x_val = diffMax(x, y)
    lpos = pos - n;
    rpos = pos + n;
    _x = x[lpos:rpos];
    _y = y[lpos:rpos];
    x_new = range(_x[1], _x[end], length=N);
    f = Spline1D(_x, _y);
    _, x_ans = diffMax(x_new, f(x_new))
    return x_ans, f(x_ans)
end
function interpx(x, y, new_y; N=10)
    pos = findfirst(y .> new_y)
    _x = x[pos-N:pos];
    _y = y[pos-N:pos];
    f = Spline1D(_y, _x);
    new_x = f(new_y);
    return new_x;
end
```

```julia
p = zeros(nr);
prob = ODEProblem(dudt!, u0, (0.0, 1.0), p);
ts, pred = get_Tcurve(prob, p; saveat=tsteps);
T_true = pred[end,:];
τ_true,_ = curvMax(tsteps, T_true);
```

```julia
Random.seed!(0x7777777);
d = Normal(0, 1);
td = truncated(d, -3, 3);

n_sample = 400;
p_sample = (rand(td, n_sample, nr)' .* log.(UF)/3)';
τ_sample = zeros(n_sample)
τT_sample = zeros(n_sample)
T_sample = zeros(n_sample, datasize)

epochs = ProgressBar(1:n_sample);
for i in epochs
    p = p_sample[i, :];
    ts, pred = get_Tcurve(prob, p; saveat=tsteps);
    T_sample[i, :] = pred[end,:];
end
```

```julia
for i in 1:n_sample
    τ_sample[i],_ = curvMax(tsteps, T_sample[i,:])
end

h1 = plot(legend=false);
plot!(tsteps, T_sample', alpha=0.2, color="#BBBBBB")
plot!(tsteps, T_true, color="#FF0000")

h2 = plot(legend=false);
density!(τ_sample, xlim=[0,tsteps[end]])

h = plot(h1, h2, layout=@layout[a{0.7h}; a{0.3h}], size=(800,600), framestyle=:box)
display(h)
```
